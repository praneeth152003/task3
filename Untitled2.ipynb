```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
np.random.seed(42)
n_samples = 100
feature1 = np.random.rand(n_samples) * 10
feature2 = np.random.rand(n_samples) * 5
noise = np.random.randn(n_samples) * 2
target = 3 * feature1 + 5 * feature2 + 10 + noise
data = pd.DataFrame({
    'Feature1': feature1,
    'Feature2': feature2,
    'Target': target
})
print("/content/Housing.csv")
print(data.head())
print("\n Dataset Description ")
print(data.describe())
print("\n Simple Linear Regression ")
X_simple = data[['Feature1']]
y_simple = data['Target']
X_train_simple, X_test_simple, y_train_simple, y_test_simple = train_test_split(
    X_simple, y_simple, test_size=0.3, random_state=42
)
simple_model = LinearRegression()
simple_model.fit(X_train_simple, y_train_simple)
y_pred_simple = simple_model.predict(X_test_simple)
mse_simple = mean_squared_error(y_test_simple, y_pred_simple)
r2_simple = r2_score(y_test_simple, y_pred_simple)
print(f"Simple Linear Regression (Feature1):")
print(f"  Coefficient: {simple_model.coef_[0]:.2f}")
print(f"  Intercept: {simple_model.intercept_:.2f}")
print(f"  Mean Squared Error (MSE): {mse_simple:.2f}")
print(f"  R-squared (R2): {r2_simple:.2f}")
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_test_simple['Feature1'], y=y_test_simple, label='Actual Data')
plt.plot(X_test_simple['Feature1'], y_pred_simple, color='red', linewidth=2, label='Regression Line')
plt.title('Simple Linear Regression (Feature1 vs. Target)')
plt.xlabel('Feature1')
plt.ylabel('Target')
plt.legend()
plt.grid(True)
plt.show()
print("\n Multiple Linear Regression ")
X_multiple = data[['Feature1', 'Feature2']]
y_multiple = data['Target']
X_train_multiple, X_test_multiple, y_train_multiple, y_test_multiple = train_test_split(
    X_multiple, y_multiple, test_size=0.3, random_state=42
)
multiple_model = LinearRegression()
multiple_model.fit(X_train_multiple, y_train_multiple)
y_pred_multiple = multiple_model.predict(X_test_multiple)
mse_multiple = mean_squared_error(y_test_multiple, y_pred_multiple)
r2_multiple = r2_score(y_test_multiple, y_pred_multiple)
print(f"Multiple Linear Regression (Feature1, Feature2):")
print(f"  Coefficients: {multiple_model.coef_}")
print(f"  Intercept: {multiple_model.intercept_:.2f}")
print(f"  Mean Squared Error (MSE): {mse_multiple:.2f}")
print(f"  R-squared (R2): {r2_multiple:.2f}")
plt.figure(figsize=(10, 6))
plt.scatter(y_test_multiple, y_pred_multiple, alpha=0.7)
plt.plot([min(y_test_multiple), max(y_test_multiple)], [min(y_test_multiple), max(y_test_multiple)],
         color='red', linestyle='--', linewidth=2, label='Perfect Prediction Line')
plt.title('Multiple Linear Regression: Actual vs. Predicted Target')
plt.xlabel('Actual Target')
plt.ylabel('Predicted Target')
plt.legend()
plt.grid(True)
plt.show()
print("\n Understanding the Results ")
print("Simple Linear Regression:")
print(f"The coefficient for Feature1 ({simple_model.coef_[0]:.2f}) indicates that for every one-unit increase in Feature1, the Target is predicted to increase by approximately {simple_model.coef_[0]:.2f} units, assuming other factors are constant (though in simple regression, there are no other features).")
print(f"The R-squared value of {r2_simple:.2f} means that approximately {r2_simple*100:.2f}% of the variance in the Target variable can be explained by Feature1 alone.")
print("\nMultiple Linear Regression:")
print(f"The coefficients are: Feature1 = {multiple_model.coef_[0]:.2f}, Feature2 = {multiple_model.coef_[1]:.2f}.")
print(f"This means for every one-unit increase in Feature1, the Target is predicted to increase by {multiple_model.coef_[0]:.2f} units, holding Feature2 constant.")
print(f"Similarly, for every one-unit increase in Feature2, the Target is predicted to increase by {multiple_model.coef_[1]:.2f} units, holding Feature1 constant.")
print(f"The R-squared value of {r2_multiple:.2f} means that approximately {r2_multiple*100:.2f}% of the variance in the Target variable can be explained by both Feature1 and Feature2 combined.")
print("\nNotice that the R-squared for multiple regression is likely higher than simple regression, indicating that including more relevant features improves the model's explanatory power.")
print("The MSE is a measure of the average squared difference between the estimated values and the actual value. Lower MSE indicates a better fit.")
```
